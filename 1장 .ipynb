1장 딥러닝이란 무엇인가? 

인공지능 이란?  보통의 사람이 수행하는 지능적인 작업을 자동화하기 위한 연구 할동.
 머신러닝 이란 ?  데이터와 해답을 가지고 머신 러닝으로 규칙을 찾아내는 것. 

데이터에서 표현을 학습하기.
머신러닝은 입력 데이터를 기반으로 기대 출력에 가깝게 만드는 유용한 표현을 학습하는 것 

데이터 처리 작업에서 규칙을 찾아 내기 위해 위 3 가지가 필요합니다.
입력 데이터 포인트 : 예를 들어 주어진 문제가 음성 인식이라면, 이 데이터 포인트는 사람의 대화가 녹음된 사운드 파일 입니다.
기대 출력 : 음성 인식 작업에서 사람이 사운드 파일을 듣고 옮긴 글, 이미지 작업에서 기대하는 출력은 '강아지', '고양이' 등과 같은 태그.
알고리즘의 성능을 측정하는 방법 : 알고리즘은 현재 출력과 기대 출력 간 차이를 결정하기 위해 필요, 측정값은 알고리즘의 작동 방식을 교정하기 위한 신호로 다시 피드백 됩니다. 이런 수정 단계를 학습이라고 말합니다.

머신 러닝에서 학습이란 ?  즉 더 나은 표현을 찾는 자동화된 과정 입니다.


딥러닝에서 '딥'이란 무엇일까? 
연속된 층으로 표현을 학습하는 개념을 나타냄.
데이터로부터 모델을 만드는 데 얼마나 많은 층을 사용하는지가 그 모델의 깊이가 됩니다.

이 분야에 대한 적절한 다른 이름은 층 기반 표현 학습, 계층적 표현 학습이 될 수 있습니다.
딥러닝에서 기본 층을 겹겹이 쌓아 올려 구성한 신경망이라는 모델을 사용하여 표현 층을 학습합니다.
